```python
#Write Code for Logistic Regression here
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Load the dataset
cancer_dataframe = pd.read_csv('/content/dataR2.csv')
# cancer_dataframe.head()                      # shows top 5 entries
# cancer_dataframe.tail()                      # shows last 5 entries
# cancer_dataframe.shape                       # shows columns x rows
# cancer_dataframe.info()                      # brief info
# cancer_dataframe.describe()                  # outputs a table of mean count max of all columns etc...
# cancer_dataframe.isnull().sum()              # shows total number of null values

cancer_dataset = cancer_dataframe.to_numpy()
# print(cancer_dataset)

X = cancer_dataset[:, :9]
Y = cancer_dataset[:, 9:].reshape(1, -1)[0]

# print(X)
# print('-----------------------------------------')
# print(Y)

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

# Logistic Regression
model = LogisticRegression(max_iter=1000)
model.fit(x_train, y_train)
print('Logistic Regression accuracy = ' , model.score(x_test, y_test))
```

SVM(Support Vector Machines)

```python
#Write Code for SVM here
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

# Load the dataset
cancer_dataframe = pd.read_csv('/content/dataR2.csv')
# cancer_dataframe.head()                      # shows top 5 entries
# cancer_dataframe.tail()                      # shows last 5 entries
# cancer_dataframe.shape                       # shows columns x rows
# cancer_dataframe.info()                      # brief info
# cancer_dataframe.describe()                  # outputs a table of mean count max of all columns etc...
# cancer_dataframe.isnull().sum()              # shows total number of null values

cancer_dataset = cancer_dataframe.to_numpy()
# print(cancer_dataset)

X = cancer_dataset[:, :9]
Y = cancer_dataset[:, 9:].reshape(1, -1)[0]

# print(X)
# print('-----------------------------------------')
# print(Y)

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

# SVM
clf = SVC(kernel='linear', C=3)
clf.fit(x_train, y_train)
print('SVM accuracy = ' , clf.score(x_test, y_test))
```

```python
#Code for Linear Regression here
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

real_estate_dataframe = pd.read_excel('/content/Real estate valuation data set.xlsx')
# real_estate_dataframe.info()                      # brief info
# real_estate_dataframe.describe()                  # outputs a table of mean count max of all columns etc...

real_estate_dataset = real_estate_dataframe.to_numpy()

X = real_estate_dataset[:, 2:7]
Y = real_estate_dataset[:, 7:]

# print(X)
# print('-----------------------------------------')
# print(Y)

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)

model = LinearRegression()
model.fit(x_train, y_train)
print('Linear Regression accuracy = ' , model.score(x_test, y_test))
```
